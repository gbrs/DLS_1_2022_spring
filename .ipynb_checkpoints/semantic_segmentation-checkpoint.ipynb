```python
class SegNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.enc_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        self.pool0 = nn.MaxPool2d(2, stride=2, return_indices=True)  # 256 -> 128
        self.enc_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.pool1 = nn.MaxPool2d(2, stride=2, return_indices=True)  # 128 -> 64
        self.enc_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.pool2 = nn.MaxPool2d(2, stride=2, return_indices=True)  # 64 -> 32
        self.enc_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU()
                         )
        self.pool3 = nn.MaxPool2d(2, stride=2, return_indices=True)  # 32 -> 16

        # bottleneck
        self.bottleneck_conv = nn.Sequential(
                                    nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(512),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(512),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(512),
                                    nn.ReLU()
                               )
        self.bottleneck_pool = nn.MaxPool2d(2, stride=2, return_indices=True)  ## 16 -> 8
        self.bottleneck_upsample = nn.MaxUnpool2d(2, stride=2)  ## 8 -> 16
        self.bottleneck_dec_conv = nn.Sequential(
                                        nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                        nn.BatchNorm2d(512),
                                        nn.ReLU(),
                                        nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                        nn.BatchNorm2d(512),
                                        nn.ReLU(),
                                        nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                                        nn.BatchNorm2d(512),
                                        nn.ReLU()
                                  )

        # decoder
        self.upsample0 = nn.MaxUnpool2d(2, stride=2)  # 16 -> 32
        self.dec_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.upsample1 = nn.MaxUnpool2d(2, stride=2)  # 32 -> 64
        self.dec_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.upsample2 = nn.MaxUnpool2d(2, stride=2)  # 64 -> 128
        self.dec_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        self.upsample3 = nn.MaxUnpool2d(2, stride=2)  # 128 -> 256
        self.dec_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=1, kernel_size=3, padding=1)  ## а где-то 2 матрицы на выходе
                         )

    def forward(self, x):
        # encoder
        e0, ind0 = self.pool0(self.enc_conv0(x))
        e1, ind1 = self.pool1(self.enc_conv1(e0))
        e2, ind2 = self.pool2(self.enc_conv2(e1))
        e3, ind3 = self.pool3(self.enc_conv3(e2))

        # bottleneck
        b0, ind4 = self.bottleneck_pool(self.bottleneck_conv(e3))
        b1 = self.bottleneck_dec_conv(self.bottleneck_upsample(b0, ind4))

        # decoder
        d0 = self.dec_conv0(self.upsample0(b1, ind3))
        d1 = self.dec_conv1(self.upsample1(d0, ind2))
        d2 = self.dec_conv2(self.upsample2(d1, ind1))
        d3 = self.dec_conv3(self.upsample3(d2, ind0))  # no activation ## про softmax?
        return d3
```

## Метрика

Используем такую метрику для оценки качества:

$I o U=\frac{\text {target } \cap \text { prediction }}{\text {target } \cup{prediction }}$


```python
def iou_pytorch(outputs: torch.Tensor, labels: torch.Tensor):
    outputs = torch.sigmoid(outputs) > 0.5

    '''squeeze() удаляет размерности размером 1.
    byte() превращает в uint8. 
    Округление до целых вниз. Нельзя после сигмоиды подавать?
    Все выходящее за пределы - 0
    '''
    outputs = outputs.squeeze(1).byte()
    labels = labels.squeeze(1).byte()
    SMOOTH = 1e-8
    intersection = (outputs & labels).float().sum((1, 2)) 
    union = (outputs | labels).float().sum((1, 2)) 
    
    iou = (intersection + SMOOTH) / (union + SMOOTH)
    
    thresholded = torch.clamp(20 * (iou - 0.5), 0, 10).ceil() / 10  # This is equal to comparing with thresolds
    
    return thresholded
```

## Функция потерь

Возьмем *бинарную кросс-энтропию*:

$$\mathcal L_{BCE}(y, \hat y) = -\sum_i \left[y_i\log\sigma(\hat y_i) + (1-y_i)\log(1-\sigma(\hat y_i))\right],$$

где $y$ это  таргет желаемого результата и $\hat y$ является выходом модели. $\sigma$ - это *логистическая* функция, преобразующая результат в диапазон $[0,1]$.

Для стабильности обучения рекомендуется использовать следующее упрощение:

$$\mathcal L_{BCE} = \hat y - y\hat y + \log\left(1+\exp(-\hat y)\right).$$

```python
def bce_loss(y_real, y_pred):
    y_pred = torch.sigmoid(y_pred)
    return (y_pred - y_real * y_pred + torch.log(1 + torch.exp(-y_pred))).sum()
```

## Функции для обучения

Создадим функцию обучения модели

```python
def train(model, opt, loss_fn, epochs, data_tr, data_val, model_name):
    X_val_view, Y_val_vew = next(iter(data_val))
    # загружаем один батч из валидации для отображения одних картинок

    train_losses = []
    val_losses = []
    # scores = [0, 0, 0, 0]  ## для UNet2 + BCE
    scores = [0, 0]
    max_score = 0

    for epoch in range(epochs):
        print('* Epoch %d/%d' % (epoch+1, epochs))
        
        # подсчет тренировочного лосса
        avg_train_loss = 0
        model.train()  # train mode
        for X_batch, Y_batch in data_tr:
            
            X_batch = X_batch.to(device)
            Y_batch = Y_batch.to(device)
            
            opt.zero_grad()
            Y_pred = model(X_batch)
            loss = loss_fn(Y_batch, Y_pred)
            loss.backward()
            opt.step()
            
            avg_train_loss += loss / len(data_tr)
            
            # откуда переполнение?
            del Y_pred, X_batch, Y_batch 
            torch.cuda.memory.empty_cache()
        
        train_losses.append(avg_train_loss.item())

        # данные для вывода картинок (промежуточных результатов)
        model.eval()
        Y_hat = model(X_val_view.to(device))
        Y_hat = Y_hat.detach().cpu()

        # считаем валидационные лоссы
        avg_val_loss = 0
        for X_batch, Y_batch in data_val:

            X_batch = X_batch.to(device)
            Y_batch = Y_batch.to(device)
            
            with torch.set_grad_enabled(False):
                Y_pred = model(X_batch)
                loss = loss_fn(Y_batch, Y_pred)

            avg_val_loss += loss / len(data_val)

            # откуда переполнение? И именно на валидации, вроде
            del Y_pred, X_batch, Y_batch 
            torch.cuda.memory.empty_cache()
        
        val_losses.append(avg_val_loss.item())

        # считаю скор на кождой эпохе
        score = score_model(model, iou_pytorch, data_val)
        scores.append(score)
        if score > max_score:
            torch.save(model.state_dict(), f'/content/gdrive/My Drive/{model_name}.pth')
            max_score = score

        # визуализация
        clear_output(wait=True)
        for k in range(6):
            plt.subplot(3, 6, k+1)
            plt.imshow(np.rollaxis(X_val_view[k].numpy(), 0, 3), cmap='gray')
            plt.title('Real')
            plt.axis('off')

            plt.subplot(3, 6, k+7)
            plt.imshow(Y_hat[k, 0], cmap='gray')
            plt.title('Output grad')
            plt.axis('off')

            # четкие маски рисуем с "уровнем отсечки" - 0.5 (после сигмоиды)
            plt.subplot(3, 6, k+13)
            plt.imshow(torch.sigmoid(Y_hat[k, 0]) > 0.5, cmap='gray')
            plt.title('Output')
            plt.axis('off')
        
        plt.suptitle('%d / %d - train loss: %f - val loss: %f - score: %f - score max: %f' 
                     % (epoch+1, epochs, avg_train_loss, avg_val_loss, score, max_score))
        plt.show()

        # Учимся минимум 20 эпох.
        # Если три последние эпохи скор на 90% ниже максимального прекращаем обучение.
        # Не слишком жестко сделал? (у других, например, лучше SegNet с BCE дальше учился)
        # Нет здесь чего-то от "проблемы подглядывания"?
        # Для UNet2 + BCE сделеал исключение: 5 подряд, а то ее быстро вышибало
        if ((epoch > 19) 
              and (0.9 * max_score > scores[epoch]) 
              and (0.9 * max_score > scores[epoch + 1]) 
              and (0.9 * max_score > scores[epoch + 2])):
            break
    
    return train_losses, val_losses, scores[4:]
```

```python
# рисователь графиков
def make_charts(train_losses, val_losses, scores, model_name):
    plt.figure(figsize=(15, 9))
    plt.plot(train_losses, label="train_loss")
    plt.plot(val_losses, label="val_loss")
    plt.title(model_name)
    plt.legend(loc='best')
    plt.xlabel("epochs")
    plt.ylabel("loss")
    plt.show()

    plt.figure(figsize=(15, 9))
    plt.plot(scores, label="score")
    plt.title(model_name)
    plt.legend(loc='best')
    plt.xlabel("epochs")
    plt.ylabel("score")
    plt.show()
```

```python
# оценка модели
def score_model(model, metric, data):
    model.eval()
    scores = 0
    for X_batch, Y_label in data:
        X_batch = X_batch.to(device)
        Y_label = Y_label.to(device)
        with torch.set_grad_enabled(False):
            Y_pred = model(X_batch)
        scores += metric(Y_pred, Y_label).mean().item()

        # чистим память ручками. Почему так?
        del Y_pred, X_batch, Y_label 
        torch.cuda.memory.empty_cache()
    
    return scores/len(data)
```

## Дополнительные функции потерь

Попробуем модель еще с тремя функциями потерь: DICE, Focal loss и Lova.

**1. Dice coefficient:**:

$$D(X,Y)=\frac{2|X\cap Y|}{|X|+|Y|}$$

Чтобы она дифференцировалась можно воспользоваться таким приближением:

$$\mathcal L_D(X,Y) = 1-\frac{\sum_i 2X_iY_i}{\sum_i(X_i+Y_i)}.$$


```python
def dice_loss(y_real, y_pred):
    SMOOTH = 1e-8
    y_pred = torch.sigmoid(y_pred)
    num = (2 * y_real * y_pred).sum()
    den =  (y_real + y_pred).sum()
    res = 1 - (num + SMOOTH) / (den + SMOOTH)
    return res
```

**2. Focal loss**  

```python
def focal_loss(y_real, y_pred, eps = 1e-8, gamma = 2):
    alpha = 1
    bce = bce_loss(y_real, y_pred)
    pt = torch.exp(-bce)
    res = alpha * torch.pow((1 - pt), gamma) * bce
    return res
```

**3. Lova loss**  

```python
def lova_loss(y_real, y_pred):
    y_pred = torch.sigmoid(y_pred)  ##

    y_pred = y_pred.flatten()
    y_real = y_real.flatten()
    signs = 2 * y_real.float() - 1
    errors = (1 - y_pred * signs)
    errors_sorted, indices = torch.sort(errors, dim=0, descending=True)
    gt_sorted = y_real[indices.data]
    # сортируем по ошибкам и предсказания, и лейблы соответственно располагаем

    # gradient
    gts = gt_sorted.sum()
    intersection = gts - gt_sorted.float().cumsum(0)
    union = gts + (1 - gt_sorted).float().cumsum(0)
    grad = 1. - intersection / union

    p = len(gt_sorted)
    grad[1:p] = grad[1:p] - grad[0:-1]
    
    loss = torch.dot(torch.relu(errors_sorted), grad)
    return loss
```

## U-Net



```python
# Создаем модель UNet.
# Без батчнормов (не упоминающихся в исходной статье) модель почему-то у меня не училась.
# Только на первом батче училась. Где я накосячил?
class UNet(nn.Module):
    def __init__(self):
        super().__init__()

        # encoder (downsampling)
        self.enc_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        self.pool0 = nn.MaxPool2d(2, stride=2)  # 256 -> 128
        self.enc_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.pool1 = nn.MaxPool2d(2, stride=2)  # 128 -> 64
        self.enc_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.pool2 = nn.MaxPool2d(2, stride=2)  # 64 -> 32
        self.enc_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU()
                         )
        self.pool3 = nn.MaxPool2d(2, stride=2)  # 32 -> 16

        # bottleneck
        self.bottleneck_conv = nn.Sequential(
                                    nn.Conv2d(in_channels=512, out_channels=1024, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(1024),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels=1024, out_channels=512, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(512),
                                    nn.ReLU()
                              )

        # decoder (upsampling)
        self.upsample0 = nn.Upsample(scale_factor=2)  # 16 -> 32
        self.dec_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=1024, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.upsample1 = nn.Upsample(scale_factor=2)  # 32 -> 64
        self.dec_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=512, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.upsample2 = nn.Upsample(scale_factor=2)  # 64 -> 128
        self.dec_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        self.upsample3 = nn.Upsample(scale_factor=2)  # 128 -> 256
        self.dec_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=1, kernel_size=1)  
                              # зачем часто делают 2 слоя на выходе?
                         )

    def forward(self, x):
        # encoder
        e0 = self.pool0(self.enc_conv0(x))
        e1 = self.pool1(self.enc_conv1(e0))
        e2 = self.pool2(self.enc_conv2(e1))
        e3 = self.pool3(self.enc_conv3(e2))

        # bottleneck
        b = self.bottleneck_conv(e3)

        # decoder
        d0 = self.dec_conv0(self.upsample0(torch.cat((b, e3), dim=1)))
        d1 = self.dec_conv1(self.upsample1(torch.cat((d0, e2), dim=1)))
        d2 = self.dec_conv2(self.upsample2(torch.cat((d1, e1), dim=1)))
        d3 = self.dec_conv3(self.upsample3(torch.cat((d2, e0), dim=1)))  # без активации
        return d3
```

Модифицируем модель путем изменения типа пулинга: заменим max-pooling на convolutions с stride=2 и upsampling на transpose-convolutions с stride=2.


```python
class UNet2(nn.Module):
    def __init__(self):
        super().__init__()

        # encoder (downsampling)
        self.enc_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        # self.pool0 = nn.MaxPool2d(2, stride=2)
        self.pool0 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, stride=2, padding=1)  # 256 -> 128
        self.enc_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.pool1 = nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, stride=2, padding=1)  # 128 -> 64
        self.enc_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.pool2 = nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, stride=2, padding=1)  # 64 -> 32
        self.enc_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU()
                         )
        self.pool3 = nn.Conv2d(in_channels=512, out_channels=512, kernel_size=3, stride=2, padding=1)  # 32 -> 16

        # bottleneck
        self.bottleneck_conv = nn.Sequential(
                                    nn.Conv2d(in_channels=512, out_channels=1024, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(1024),
                                    nn.ReLU(),
                                    nn.Conv2d(in_channels=1024, out_channels=512, kernel_size=3, padding=1),
                                    nn.BatchNorm2d(512),
                                    nn.ReLU()
                              )

        # decoder (upsampling)
        self.upsample0 = nn.ConvTranspose2d(in_channels=1024, out_channels=1024, kernel_size=3, stride=2, padding=1, output_padding=1)  # 16 -> 32
        # такое сочетание позволяет получить 32, а не 31 или 33 на выходе.
        self.dec_conv0 = nn.Sequential(
                              nn.Conv2d(in_channels=1024, out_channels=512, kernel_size=3, padding=1),
                              nn.BatchNorm2d(512),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=512, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU()
                         )
        self.upsample1 = nn.ConvTranspose2d(in_channels=512, out_channels=512, kernel_size=3, stride=2, padding=1, output_padding=1)  # 32 -> 64
        self.dec_conv1 = nn.Sequential(
                              nn.Conv2d(in_channels=512, out_channels=256, kernel_size=3, padding=1),
                              nn.BatchNorm2d(256),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=256, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU()
                         )
        self.upsample2 = nn.ConvTranspose2d(in_channels=256, out_channels=256, kernel_size=3, stride=2, padding=1, output_padding=1)  # 64 -> 128
        self.dec_conv2 = nn.Sequential(
                              nn.Conv2d(in_channels=256, out_channels=128, kernel_size=3, padding=1),
                              nn.BatchNorm2d(128),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=128, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU()
                         )
        self.upsample3 = nn.ConvTranspose2d(in_channels=128, out_channels=128, kernel_size=3, stride=2, padding=1, output_padding=1)  # 128 -> 256
        self.dec_conv3 = nn.Sequential(
                              nn.Conv2d(in_channels=128, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3, padding=1),
                              nn.BatchNorm2d(64),
                              nn.ReLU(),
                              nn.Conv2d(in_channels=64, out_channels=1, kernel_size=1)  
                         )

    def forward(self, x):
        # encoder
        e0 = self.pool0(self.enc_conv0(x))
        e1 = self.pool1(self.enc_conv1(e0))
        e2 = self.pool2(self.enc_conv2(e1))
        e3 = self.pool3(self.enc_conv3(e2))

        # bottleneck
        b = self.bottleneck_conv(e3)

        # decoder
        d0 = self.dec_conv0(self.upsample0(torch.cat((b, e3), dim=1)))
        d1 = self.dec_conv1(self.upsample1(torch.cat((d0, e2), dim=1)))
        d2 = self.dec_conv2(self.upsample2(torch.cat((d1, e1), dim=1)))
        d3 = self.dec_conv3(self.upsample3(torch.cat((d2, e0), dim=1)))  # без активации
        return d3
```
